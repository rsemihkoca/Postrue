```python
from sqlalchemy import create_engine, Column, Integer, String, Boolean, Float, DateTime
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
import datetime
import random

Base = declarative_base()

class Transactions(Base):
    __tablename__ = 'Transactions'

    Datetime = Column(DateTime, primary_key=True)
    Year = Column(Integer, primary_key=True)
    Month = Column(Integer, primary_key=True)
    Day = Column(Integer, primary_key=True)
    ClientId = Column(Integer, primary_key=True)
    ClientName = Column(String)
    BadPosture = Column(Boolean)
    TorsoInclination = Column(Float)
    NeckInclination = Column(Float)
    Result = Column(String)
    Image = Column(String)

POSTGRESQL_CON_STR = "postgresql+psycopg2://postgres:postgres@127.0.0.1:5432/postgres"
engine = create_engine(POSTGRESQL_CON_STR)
Session = sessionmaker(bind=engine)
session = Session()

start_date = datetime.datetime.now() - datetime.timedelta(days=60)
current_date = start_date

data_to_insert = []

while current_date <= datetime.datetime.now():
    num_data_points = random.randint(0, 10)
    
    for _ in range(num_data_points):
        neck_inclination = random.uniform(1, 80)
        torso_inclination = random.uniform(1, 26)
        
        bad_frames = 0
        if neck_inclination >= 40 or torso_inclination >= 10:
            bad_frames += 1
        
        if bad_frames == 0:
            result = 'Good'
        else:
            if neck_inclination > 40 and torso_inclination < 13:
                result = 'Bad Neck'
            elif neck_inclination < 40 and torso_inclination > 13:
                result = 'Bad Torso'
            else:
                result = 'Bad NeckTorso'

        data = {
            "Datetime": current_date,
            "Year": current_date.year,
            "Month": current_date.month,
            "Day": current_date.day,
            "ClientId": 1,
            "ClientName": "admin",
            "BadPosture": bad_frames > 0,
            "TorsoInclination": torso_inclination,
            "NeckInclination": neck_inclination,
            "Result": result,
            "Image": "https://i.imgur.com/2Yj2Q8k.jpg"
        }
        
        data_to_insert.append(data)
        
    current_date += datetime.timedelta(days=1)

for data in data_to_insert:
    transaction = Transactions(**data)
    session.add(transaction)

session.commit()
session.close()
```
